Distiluse-base-multilingual-cased-v1 - sentence-transformers model: maps sentences and paragraphs to a 512 dimensional dense vector space. It can be used for tasks like clustering or semantic search. Captures semantic meaning.

```python
!pip install sentence-transformers
from langchain_community.embeddings import HuggingFaceEmbeddings
modelPath = "distiluse-base-multilingual-cased-v1" # Pre-trained model provided by sentence-transformers

model_kwargs = {'device':'cuda'} # Sets model location to cuda

encode_kwargs = {'normalize_embeddings':False} #Doesn't normalize, should be?

# Helps optimize similarity scores

# HuggingFaceEmbeddings: embeddings (still a bit unsure specifically about HuggingFace)
embeddings = HuggingFaceEmbeddings(
    model_name = modelPath,
    model_kwargs = model_kwargs,
    encode_kwargs = encode_kwargs
)
```

```python
# FAISS - helps semantic search, also uses nearest-neighbor to find most similar cluster based on query vectors provided
# Semantic Search - Takes user's query, and goes past keyword matching, but understands semantics (meaning), returns related results
# Uses Transformers, extracts dense vectors that contain semantic information, and efficiently does similarity matching as well.
from langchain.vectorstores import FAISS
```

```python
# from_documents FAISS(another e.g. Chroma) - so we can search for similarities, we initialize a vector space
# initializes a database where data is data points, embeddings are corresponding vector representations
!pip install faiss-gpu
db = FAISS.from_documents(data, embedding = embeddings)
```

```python
# Used 7B Llama, but limiations of GPU
# To do Korean, use skt's kogpt
# 125M parameter nums
# Pipeline - does the text's tokenize, forward, and post processing all at once
# Pipeline - can use all HuggingFace models, but when there are strict guidelines, it's not good
from langchain_community.llms.huggingface_pipeline import HuggingFacePipeline
from transformers import GPT2LMHeadModel,AutoTokenizer,pipeline
model_id = "skt/ko-gpt-trinity-1.2B-v0.5" # specifies pre-trained model, which is a Korean GPT-2 model
tokenizer = AutoTokenizer.from_pretrained(model_id) # Loads tokenizer corresponding to specified model_id - responsible for converting text into format that model can process
model = GPT2LMHeadModel.from_pretrained(model_id).to(device) # Loads GPT model to model_id
pipe = pipeline("text-generation",model = model, tokenizer = tokenizer, max_new_tokens=200,device=device) # Creates text generation pipeline
hf = HuggingFacePipeline(pipeline = pipe) # Wrapping?
```

**RAG with LCEL(LangChain Expression Language)**

LangChain provides LCEL so it can implement complicated chains. LangChain's objects with Runnable Interface contain the methods invoke, batch, stream, etc. It then connects these objects with the pipe operator | and creates one chain. These connected objects take the previous object's outputs and takes it as the input.

**invoke**

Calls chain about the input. In this case, it's a single chain.

**streaming**

If the model's size or prompt is too big the response time can be slow. It waits for all the responses and then outputs a response, but this can take a long time, so LangChain suupports a streaming pattern. This means as outputs are generated, it can return responses in real time through streaming.

**batch**

If there are a lot of questions or commands, instead of running a loop, it stores the prompts in a list, and through batch it can take care of multiple questions at the same time.

```python
from langchain_core.prompts import PromptTemplate
# Takes the strings made above and uses the class PromptTemplate to dynamically fill it with context and question sections to combine with new prompts
# .from_template : Takes string and generates it as a PromptTemplate Instance
# .format(context = "~~", question = "~~") : Fills template to make a prompt

template = """마지막에 질문에 답하려면 다음과 같은 맥락을 사용합니다.

{context}

질문: {question}

유용한 답변:"""

custom_rag_prompt = PromptTemplate.from_template(template)
custom_rag_prompt.invoke({"context":"Apple is sweet.","question" : "How apple taste?"})
```

```python
from langchain_core.prompts import PromptTemplate # creates and manages prompt templates
from langchain_core.runnables import RunnablePassthrough #simply passes input through
from langchain_core.output_parsers import StrOutputParser #parses output into human-readable string format

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs) # Takes list of document objects and joins their 'page_content' attributes into a single string

rag_chain = (
    # Takes documents that retriever searched (4 of them) and joins them to "\n\n" through format_docs
    # RunnablePassthrough() : Relays inputted value
    # First chain - form of {"context" : format_docs, "question" : 질문}, dictionary goes into next chain as input
    {"context": retriever | format_docs, "question": RunnablePassthrough()} # retriever fetches documents, format_docs processes them (| used to chain)
    # Custom_rag_prompt takes dictionary and gneerates completed prompt string
    | custom_rag_prompt
    # passed to Hugging Face pipeline ('hf') which processes the prompt and generates a response
    | hf
    # Parses model's output to a form that people can read.
    | StrOutputParser()
)
```

```python
# Tried to put it into GPT Neo - dimension issue because of Korean
# 도배지에 녹은 자국이 발생하는 주된 원인과 그 해결 방법은 무엇인가요?  Random?
# Solution: Use KoGPT in an above cell
## Note> We use a small model, so reached limit to possible generated tokens, so address pipeline's max_new_tokens
tmp_list = []
for chunk in rag_chain.stream("도배지에 녹은 자국이 발생하는 주된 원인과 그 해결 방법은 무엇인가요?"): # Iterates over each chunk of text generated by rag_chain
    tmp_list.append(chunk) # Appends each generated chunk of text to tmp_list - collect all parts of the response
    print(chunk, end = "", flush = True) # Use print's flush = True to empty output buffer, each chunk is printed as soon as it's generated
# print(tmp_list)
```
